```python
def get_tags(tag,batch_sent):
    _tags = list()
    batch_sent = batch_sent.detach().numpy()
    print(type(batch_sent))
    for row in range(len(batch_sent)):
        for col in range(len(batch_sent[row])):
            if tag[row][col] not in [0,1,2]:
                ix = batch_sent[row][col]
                print(ix)
                _tags.append(word2id[ix])
    return _tags
```
